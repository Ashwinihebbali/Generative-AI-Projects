```python
from google import genai
from getpass import getpass
api_key=getpass("enter your gemini API Key:")
client=genai.Client(api_key=api_key)
```

```python
print("AI RESEARCH PAPER SIMPLIFIER")
print("-" * 50)

research_text = input("Paste the research paper paragraph here:\n")
level = input("Explanation level (Beginner / Student / Technical): ")
length = input("Explanation length (Short / Medium / Detailed): ")
```

```python
prompt = f"""
You are an AI assistant that simplifies research papers.

Research Text:
{research_text}

Explain this research content in a simple way.

Explanation Level: {level}
Length: {length}

Instructions:
1. Use simple language.
2. Avoid complex terminology.
3. Explain the key idea clearly.
4. Make it easy for students to understand.
"""
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print("\nSIMPLIFIED EXPLANATION")
print("-" * 50)
print(response.text)
```
